```python
# Step 1: Filter the DataFrame to include only rows where a medal was won
medals_df = df.dropna(subset=['Medal'])

# Step 2: Group by 'Team' and 'Medal' to count the number of each type of medal
medal_counts = medals_df.groupby(['Team', 'Medal']).size().unstack(fill_value=0)

# Step 3: Plotting the medal counts
# Create a bar plot for each medal type

plt.figure(figsize=(14, 8))

# Function to add labels on top of bars
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], str(y[i]), ha='center', va='bottom')

# Plotting Gold medals
plt.subplot(3, 1, 1)
gold_medals = medal_counts['Gold'].sort_values(ascending=False).head(10)
gold_medals.plot(kind='bar', color='gold')
add_labels(gold_medals.index, gold_medals.values)
plt.title('Top 10 Countries by Gold Medals')
plt.xlabel('Country')
plt.ylabel('Gold Medals')

# Plotting Silver medals
plt.subplot(3, 1, 2)
silver_medals = medal_counts['Silver'].sort_values(ascending=False).head(10)
silver_medals.plot(kind='bar', color='silver')
add_labels(silver_medals.index, silver_medals.values)
plt.title('Top 10 Countries by Silver Medals')
plt.xlabel('Country')
plt.ylabel('Silver Medals')

# Plotting Bronze medals
plt.subplot(3, 1, 3)
bronze_medals = medal_counts['Bronze'].sort_values(ascending=False).head(10)
bronze_medals.plot(kind='bar', color='#cd7f32')
add_labels(bronze_medals.index, bronze_medals.values)
plt.title('Top 10 Countries by Bronze Medals')
plt.xlabel('Country')
plt.ylabel('Bronze Medals')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Filter the DataFrame to include only rows where a medal was won
medals_df = df.dropna(subset=['Medal'])

# Step 2: Group the data by Gender and count the medals
gender_medal_counts = medals_df['Gender'].value_counts()

# Step 3: Plotting the male and female medal ratio with the updated approach
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_medal_counts.index, y=gender_medal_counts.values, hue=gender_medal_counts.index, palette='viridis', legend=False)

# Adding the exact counts on top of the bars
for i, count in enumerate(gender_medal_counts.values):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.title('Male vs Female Medal Ratio')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.show()
```
